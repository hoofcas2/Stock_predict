```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mms=MinMaxScaler()
ss_scaled=mms.fit_transform(samsung[['Open', 'High', 'Low', 'Close', 'Volume']])
```

```python
samsung.columns
```

```python
ss_scaled=pd.DataFrame(ss_scaled)
ss_scaled.columns=['Open', 'High', 'Low', 'Close', 'Volume']
```

```python
ss_scaled
```

```python
#학습 데이터 : 과거~ 300일 이전 까지의 데이터
#테스트 데이터 : 300일 전부터 현재까지
```

```python
train=ss_scaled[:-300] #5700건
test=ss_scaled[-300:] #300건
```

```python
#window_size를 30으로 설정, 30일 동안의 주가 데이터(xdata) -> 다음 날 종가(ydata) 예측
```

```python
train.columns
train_feature=train[['Open', 'High', 'Low', 'Volume']]
```

```python
train_label=train[['Close']]
```

```python
train_feature
```

```python
train_label
```

```python
#train_feature, train_label, 윈도우 크기 : 30
window_size=30
```

```python
#훈련 데이터를 윈도우 크기에 따라 재구성
```

```python
feature_list=[]
label_list=[]
```

```python
for i in range(len(train_feature)-window_size): #0 ~ 5670-1까지 반복
    feature_list.append(np.array(train_feature.iloc[i:i+window_size]))
    label_list.append(np.array(train_label.iloc[i+window_size]))  
    
```

```python
feature_list=np.array(feature_list)
label_list=np.array(label_list)
```

```python
len(feature_list) #5670개
```

```python
from sklearn.model_selection import train_test_split
```

```python
#train, valid set 생성
xtrain,xvalid,ytrain,yvalid=train_test_split(feature_list, label_list, test_size=0.2)
```

```python
np.shape(xtrain) #(4536, 30, 4)
np.shape(ytrain) #(4536, 1)

np.shape(xvalid) #(1134, 30, 4)
np.shape(yvalid) #(1134, 1)
```

```python
xtrain.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM
from keras.callbacks import EarlyStopping, ModelCheckpoint
```

```python
test_feature=test[['Open', 'High', 'Low', 'Volume']]
test_label=test[['Close']]
```

```python
feature_test=[]
label_test=[]
for i in range(len(test_feature)-window_size): 
    feature_test.append(np.array(test_feature.iloc[i:i+window_size]))
    label_test.append(np.array(test_label.iloc[i+window_size]))      
```

```python
feature_test=np.array(feature_test)
label_test=np.array(label_test)
```

```python
#LSTM모델
```

```python
model=Sequential()
```

```python
model.add(LSTM(16,input_shape=(np.shape(xtrain)[1], np.shape(xtrain)[2]),
               return_sequences=False))
```

```python
model.add(Dense(1))
```

```python
model.compile(loss='mse', optimizer='adam')
```

```python
es=EarlyStopping(monitor='val_loss', patience=30)
cp=ModelCheckpoint('tmp_cp.h5', monitor='val_loss', save_best_only=True)
```

```python
model.fit(xtrain, ytrain, epochs=100, batch_size=16, 
          validation_data=(xvalid, yvalid), callbacks=[es,cp])
```

```python
pred=model.predict(feature_test)
```

```python
pred
label_test
plt.figure(figsize=(12,10))
plt.plot(pred, label='prediction')
plt.plot(label_test, label='actual')
plt.legend()
```
